```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('books.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.isna().any()
#mengecek apakah ada data kosong
#jika True berarti ada
#jika False berarti tidak ada
```

```python
#hapus missing value
df = df.dropna()
```

```python
df.isna().any()
```

```python
#hapus kolom index, sales rank dan author rating
df = df.drop('index', axis=1)
df = df.drop('sales rank', axis=1)
df = df.drop('Author_Rating', axis=1)
df.head()
```

```python
#supaya tidak ada.0 di publishing year
df['Publishing Year'] = df['Publishing Year'].fillna(0).astype(int)
df['Book_ratings_count'] = df['Book_ratings_count'].fillna(0).astype(int)
df['units sold'] = df['units sold'].fillna(0).astype(int)
df.head()
```

```python
df.info()
```

```python
df['genre'].value_counts(dropna=False)
```

```python
df['genre'] = df['genre'].str.strip()
df['genre'] = df['genre'].str.replace('genre', '', case=False, regex=False)
```

```python
df.head()
```

```python
df['genre'].value_counts(dropna=False)
```

```python
df.isnull().sum()
```

```python
df.to_csv('very_clean.csv', index=False)

from google.colab import files
files.download('very_clean.csv')
```
